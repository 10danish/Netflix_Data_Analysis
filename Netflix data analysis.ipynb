```python
data=pd.read_csv("netflix_ratings.csv")
data=data.dropna()
```

```python
data_filtered = data[data['type'].isin(['MOVIE', 'SHOW'])]

# Calculate average IMDb score every year
avg_score_per_year = data_filtered.groupby('release_year')['imdb_score'].mean()
year = avg_score_per_year.index.to_numpy()  # Extract years
avg_score = avg_score_per_year.to_numpy()  # Extract average scores

plt.figure(figsize=(10, 4))  
plt.scatter(year, avg_score)

plt.xlabel("Release Year")
plt.ylabel("Average IMDb Score")

plt.title("Average IMDb Score by Release Year (Movies/Shows)")

plt.grid(True)  
plt.show()
```

```python
df2=df2.drop(columns=['id','imdb_id','description'])
```

```python
df2
```

```python
df2.info()
```

```python
cat_cols=['title','type','age_certification']
le=LabelEncoder()
for i in cat_cols:
    df2[i]=le.fit_transform(df2[i])
df2.dtypes
```

```python
df2.columns
```

```python
skewed_features=['index','title', 'type', 'release_year', 'age_certification',
       'runtime', 'imdb_votes']
for i in skewed_features:
    df2[i]=np.log(df2[i]+1)
```

```python
X=df2.drop(labels=['imdb_score','index'],axis=1)
Y=df2['imdb_score']
X.head()
```

```python
Y.head(10)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
lm=LinearRegression()
lm=lm.fit(X_train,Y_train)
train_pred=lm.predict(X_train)
test_pred=lm.predict(X_test)
RMSE_test=np.sqrt(mean_squared_error(Y_test,test_pred))
RMSE_train=np.sqrt(mean_squared_error(Y_train,train_pred))
print("RMSE TrainingData = ",str(RMSE_train))
print("RMSE TestData = ",str(RMSE_test))
print('-'*50)
print("RSquared value on train: ",lm.score(X_train,Y_train))
print("RSquared value on test: ",lm.score(X_test,Y_test))
```

```python
errors=abs(test_pred-Y_test)
mape=100*(errors / Y_test)
accuracy=100-np.mean(mape)
print("Accuracy:",round(accuracy,2),'%')
```
